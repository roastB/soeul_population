[목표.1] 서울 생활인구를 DataFrame으로 만들어보자. 그리고 이후 활용해보자!

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 서울 열린데이터광장에서 다운로드 받은 파일 가져오기
raw_data = pd.read_csv("/home/cho/pp_ws/data/seoul_population_2024.csv", encoding="cp949")
raw_data

# UnicodeDecodeError: 'utf-8' codec can't decode byte 0xc0 in position 0: invalid start byte 에러코드!
# -> 인코딩 과정에서의 에러[encoding="cp949"추가]
```

```python
# 기준일(240823) - 시군구 - 총생활인구 - 내국인생활인구 - 일최대인구 - 일최소인구

df = pd.DataFrame({"Latest_update" : raw_data["기준일ID"],
                   "Gu" : raw_data["시군구명"],
                  "Total_population" : raw_data["총생활인구수"],
                  "Local" : raw_data["내국인생활인구수"],
                  "Foreign_Long" : raw_data["장기체류외국인인구수"],
                  "Foreign_Short" : raw_data["단기체류외국인인구수"]})

df
```

```python
# 최신데이터(240823)만 가져오기
df = df[ df["Latest_update"] == 20240823 ]
df
```

```python
# 디테일 수정
df.drop(index=26, inplace=True)
df.drop(labels="Latest_update", axis=1, inplace=True)

df
```

```python
#set_index()
df.set_index("Gu", inplace=True)
df
```

```python
df.info()
```

```python
# [1차 목표] : 자치구별 서울 총생활인구중 내국인생활인구/외국인 비율(proportion)
# (1) propportion data
local_prop = df["Local"] / df["Total_population"] * 100
local_prop
```

```python
df["Domestic_ratio"] = local_prop
df
```

```python
total_foreign = df["Foreign_Long"] + df["Foreign_Short"]
total_foreign_prop = total_foreign / df["Total_population"] * 100
total_foreign_prop
```

```python
df["total_Foreigner_ratio"] = total_foreign_prop
df
```

```python
# 확인!
df["Domestic_ratio"] + df["total_Foreigner_ratio"]
```

```python
# [2차 목표] : ***정규화 데이터 (normalization)
# 정규화된 데이터들을 다시 정리해서 저장!!
# crime_gu_norm = crime_gu[occur_col] / crime_gu[occur_col].max() : 최대수치에서 기본값을 나눈것이 정규화데이터!

local_norm = df["Domestic_ratio"] / 100
total_foreign_norm = df["total_Foreigner_ratio"] / 100

# 🟨 지금은 야매로 추후에... 정규화 모듈 찾아보기
```

```python
L = round(local_norm, 3)
L
```

```python
total_F = round(total_foreign_norm, 3)
total_F
```

```python
# 확인
L + total_F
```

```python
# DataFrame에 저장
df["Local_norm"] = L
df["total_Foreign_norm"] = total_F

df
```

```python
# 일단 한번 "저장"
df.to_csv("/home/cho/pp_ws/data/seoul_pop_norm.csv", sep=",", encoding="utf-8")
```

[목표.2] 서울 생활인구를 DataFrame인 "seoul_pop_norm.csv"을 통해 folium을 이용하여  자치구별 생활인구를 표시해보자.

```python
# folium을 이용하여  자치구별 Map 정보가져오기
# geo data -> Lucy Park 선생님의 서울맵을 github에서 불러와 사용.

import folium
import json

soeul_pop = pd.read_csv("/home/cho/pp_ws/data/seoul_pop_norm.csv")

geo_path = "/home/cho/pp_ws/data/02. skorea_municipalities_geo_simple.json"
geo_str = json.load(open(geo_path, encoding="utf-8")) #json.load() : json -> python file (struct)
```

```python
soeul_pop
```

```python
# [1]"자치구"별 서울 총생활인구중 "내국인생활인구"

seoul_m = folium.Map(location = [37.5502,126.982], zoom_start = 11.4)

# Choropleth map
folium.Choropleth(
    geo_data = geo_str, # geo_struct의 줄임말
    data = soeul_pop,
    #참고 : https://python-visualization.github.io/folium/latest/getting_started.html
    columns = ["Gu", "Local_norm"], # 🚩주의! 꼭 soeul_pop를 확인해줄것!
    key_on = "properties.name", # 🚩주의! 꼭 geo_str를 확인해줄것!
    fill_color = "PuRd",
    fill_opacity = 0.7,
    line_opacity = 0.2,
    legend_name=""" "자치구"별 서울 총생활인구중 내국인생활인구 "비율" """
    ).add_to(seoul_m)

seoul_m
```

```python
# [2]"자치구"별 서울 총생활인구중 "총외국인생활인구(장기체류+단기체류)""

seoul_m = folium.Map(location = [37.5502,126.982], zoom_start = 11.4)

# Choropleth map
folium.Choropleth(
    geo_data = geo_str, # geo_struct의 줄임말
    data = soeul_pop,
    #참고 : https://python-visualization.github.io/folium/latest/getting_started.html
    columns = ["Gu", "total_Foreign_norm"], # 🚩주의! 꼭 soeul_pop를 확인해줄것!
    key_on = "properties.name", # 🚩주의! 꼭 geo_str를 확인해줄것!
    fill_color = "PuRd",
    fill_opacity = 0.7,
    line_opacity = 0.2,
    legend_name=""" "자치구"별 서울 총생활인구중 내국인생활인구 "비율" """
    ).add_to(seoul_m)

seoul_m
```
